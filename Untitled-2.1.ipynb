```python
# -*- coding: utf-8 -*-
# K-Means con scikit-learn + grÃ¡ficas (Elbow, Silhouette y Clusters)

from pathlib import Path
import tempfile
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# ================== 1) Rutas y opciones ==================
DATA_PATH = Path("/Users/victorianaomivilchismontalvo/Documents/Analitica/Mall_Customers.csv")
RANDOM_STATE = 42
DROP_GENDER = True  # Si quieres incluir Gender como 0/1, pon False

def carpeta_escribible(candidatos):
    for p in candidatos:
        try:
            p.mkdir(parents=True, exist_ok=True)
            t = p / ".perm_test"
            t.write_text("ok")
            t.unlink()
            return p
        except Exception:
            continue
    raise OSError("No encontrÃ© carpeta escribible")

OUTPUT_DIR = carpeta_escribible([
    Path("/Users/victorianaomivilchismontalvo/Documents/Analitica"),
    Path("/Users/victorianaomivilchismontalvo/Downloads"),
    Path(tempfile.gettempdir()),
])
print(f"ðŸ“‚ GuardarÃ© resultados en: {OUTPUT_DIR}")

# ================== 2) Carga y limpieza ==================
def normalizar_columnas(df):
    df.columns = (df.columns.astype(str).str.strip()
                  .str.replace(r"\s+", "_", regex=True)
                  .str.replace(r"[^0-9A-Za-z_]+", "", regex=True))
    return df

def cargar_mall_customers(path: Path) -> pd.DataFrame:
    if not path.exists():
        raise FileNotFoundError(f"No se encontrÃ³ el archivo: {path}")
    df = pd.read_csv(path)
    df = normalizar_columnas(df)
    # Renombres tÃ­picos
    rename = {}
    for c in df.columns:
        lc = c.lower()
        if lc in ["annual_income_k","annualincome","annual_income_ks"]:
            rename[c] = "Annual_Income_k"
        if "spending" in lc:
            rename[c] = "Spending_Score"
        if lc == "customerid": rename[c] = "CustomerID"
        if lc == "gender":     rename[c] = "Gender"
        if lc == "age":        rename[c] = "Age"
    df = df.rename(columns=rename)
    # Quitar IDs
    for col in ["CustomerID","Unnamed_0","id","ID"]:
        if col in df.columns:
            df = df.drop(columns=col)
    # Codificar gÃ©nero
    if "Gender" in df.columns:
        df["Gender"] = (df["Gender"].astype(str).str.strip().str.lower()
                        .map({"male":1,"m":1,"female":0,"f":0}))
    return df

df = cargar_mall_customers(DATA_PATH)

# ================== 3) SelecciÃ³n y escalado ==================
features = ["Age","Annual_Income_k","Spending_Score"]
if (not DROP_GENDER) and ("Gender" in df.columns):
    features = ["Gender"] + features

X = df[features].dropna().copy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X.values)

# ================== 4) Elegir k (Elbow + Silhouette) ==================
ks = range(2, 11)
inertias, silhouettes = [], []
for k in ks:
    km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
    labels_tmp = km.fit_predict(X_scaled)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_scaled, labels_tmp))

k_opt = ks[int(np.argmax(silhouettes))]
print(f"âœ… k Ã³ptimo por silhouette: {k_opt}")
print("Inertia:", dict(zip(ks, np.round(inertias,2))))
print("Silhouette:", dict(zip(ks, np.round(silhouettes,3))))

# ================== 5) K-Means final y centros ==================
kmeans = KMeans(n_clusters=k_opt, random_state=RANDOM_STATE, n_init=10)
labels = kmeans.fit_predict(X_scaled)

centers_original = scaler.inverse_transform(kmeans.cluster_centers_)
centers_df = pd.DataFrame(centers_original, columns=features)
centers_df.index.name = "Cluster"

# TamaÃ±os por cluster
counts = pd.Series(labels).value_counts().sort_index()
counts.name = "n"
centers_df.insert(0, "n", counts.values)

print("\n=== CENTROS (escala original) ===")
print(centers_df.round(2))

# ================== 6) Guardar CSVs ==================
centers_path = OUTPUT_DIR / "kmeans_centers.csv"
labels_path  = OUTPUT_DIR / "kmeans_labels.csv"

centers_df.round(4).to_csv(centers_path, index=True, encoding="utf-8")
X.assign(cluster=labels).to_csv(labels_path, index=False, encoding="utf-8")

print("\nâœ… Archivos guardados en:")
print(f" - {centers_path}")
print(f" - {labels_path}")

# ================== 7) GRÃFICAS ==================
sns.set(style="whitegrid")

# 7.1 Elbow
plt.figure(figsize=(6,4))
plt.plot(ks, inertias, marker='o')
plt.title("MÃ©todo del Codo (Elbow)")
plt.xlabel("NÃºmero de clusters (k)")
plt.ylabel("Inercia")
plt.grid(True)
plt.tight_layout()
plt.savefig(OUTPUT_DIR / "elbow.png", dpi=150)
plt.show()

# 7.2 Silhouette
plt.figure(figsize=(6,4))
plt.plot(ks, silhouettes, marker='o')
plt.title("Puntaje Silhouette por k")
plt.xlabel("NÃºmero de clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.tight_layout()
plt.savefig(OUTPUT_DIR / "silhouette.png", dpi=150)
plt.show()

# 7.3 Scatter de clusters finales (Ingreso vs Gasto)
if {"Annual_Income_k","Spending_Score"}.issubset(set(features)):
    plt.figure(figsize=(7,5))
    sns.scatterplot(
        x=X["Annual_Income_k"], y=X["Spending_Score"],
        hue=labels, palette="tab10", s=60, alpha=0.8, legend=True
    )
    # Centroides
    plt.scatter(
        centers_df["Annual_Income_k"], centers_df["Spending_Score"],
        c="black", s=160, marker="X", label="Centros"
    )
    plt.title(f"AgrupaciÃ³n K-Means (k={k_opt})")
    plt.xlabel("Ingreso anual (k)")
    plt.ylabel("Spending Score")
    plt.legend()
    plt.tight_layout()
    plt.savefig(OUTPUT_DIR / "clusters_scatter.png", dpi=150)
    plt.show()
```
